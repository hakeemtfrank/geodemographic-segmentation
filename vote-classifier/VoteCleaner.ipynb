```python
vote_df['Republican'] = None
```

```python
dem_col = vote_df['spend_zips_poli_dem']
rep_col = vote_df['spend_zips_poli_rep']
```

```python
dem_col[2] > rep_col[2]
```

```python
for e in dem_col:
    print(dem_col)
```

```python
row = range(0, len(vote_df))
for e in row:
    if rep_col[e] > dem_col[e]:
        vote_df['Republican'].iloc[e] = 1
    else:
        vote_df['Republican'].iloc[e] = 0
```
